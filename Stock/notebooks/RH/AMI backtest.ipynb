```python
import pandas as pd
```

```python
def read_results(file):
    ami_trades = pd.read_excel(file, index_col='Date', parse_dates=True)
    ami_trades.drop(['Trade', '% chg', 'Profit', 'Shares', 'Cum. Profit', '# bars', 'Profit/bar', 'MAE', 'MFE', 'Scale In/Out'], axis=1, inplace=True)
    return ami_trades
```

```python
ami_trades = read_results('SPY_20160527.xlsx')
ami_trades.head()
```

```python
ami_trades = read_results('SPY_1999_2017_results.xls')
ami_trades
```

```python
spy_data.rename(columns={'Date/Time':'Date', 'O': 'Open', 'H': 'High', 'L': 'Low', 'C': 'Close', 'V': 'Volume'}, inplace=True)
spy_data.drop('Ticker', axis=1, inplace=True)
spy_data.set_index('Date', inplace=True)
```
