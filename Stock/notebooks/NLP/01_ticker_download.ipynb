```python
import requests as req
from bs4 import BeautifulSoup
from datetime import datetime
import os
```

```python
def get_all_for_ticker(query, page):
    url = 'https://newsapi.org/v2/everything?q={}'\
          '&language=en&from=2010-1-1'\
          '&sortBy=publishedAt'\
          '&sources=reuters,bloomberg'\
          '&page={}'.format(query, page)
    resp = req.get(url, headers={'X-Api-Key':'2dc4977049f74dda9e75c4c11f938cad',
                                 'X-No-Cache':'false'})
    return resp.json()
```

```python
def process_url(url):
    resp = req.get(url)
    resp_text = resp.text
    soup = BeautifulSoup(resp_text)
    l = list(map(lambda x: x.get_text(), soup.findAll('p')))
    return ' '.join(l)
```
